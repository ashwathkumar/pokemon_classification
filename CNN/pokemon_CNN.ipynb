# Find your starter pokemon

Here I am using CNNs to make a makeshift pokemon classifier. To manage the dataset size and the time to train the model, I will be using only the 3 starter pokemon and pikachu (cause who doesnt love pikachu) - charmandar, squirtle, bulbasaur and pikachu. Firstly, importing all libs. 

```python
import numpy as np
import pandas as pd
import keras
import os
import re
import random
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
from keras.layers import Input,Dense,Flatten,Dropout,merge,Reshape,Conv2D,MaxPooling2D,UpSampling2D,Conv2DTranspose
from keras.layers.normalization import BatchNormalization
from keras.models import Model,Sequential,load_model
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adadelta, RMSprop,SGD,Adam
from keras import regularizers
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.layers.core import Activation
from PIL import Image
from keras.applications.imagenet_utils import preprocess_input
import glob
import pickle
```

I separated the dataset into training and validation using the code below before working on the model.

```python
#Make the directories. All images were originally in a directory called pokemon. 
os.makedirs("dataset/training")
os.makedirs("dataset/validation")
for dirs in os.listdir('./dataset/dataset'):
    os.makedirs("dataset/training/"+dirs)
    os.makedirs("dataset/validation/"+dirs)
```

```python
os.listdir('./dataset/dataset')
```

Function to get list of all files. 

```python
def getListOfFiles(dirName):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = os.listdir(dirName)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(dirName, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)
                
    return allFiles
```

Get all files and separate them into respective directories. I am using a 90% train and 10% validation split

```python
allFiles = getListOfFiles('dataset/')
allFiles.sort()
random.seed(42)
random.shuffle(allFiles)
split_1 = int(0.9 * len(allFiles))
train_filenames = allFiles[:split_1]
val_filenames = allFiles[split_1:]
tmp = [os.rename (file,file.replace("dataset\\","training\\")) for file in train_filenames]
tmp = [os.rename (file,file.replace("dataset\\","validation\\")) for file in val_filenames]
```

```python
print ("Total number of images ="+ str(len(allFiles)))
print ("Training ="+ str(len(train_filenames)))
print ("Validation ="+ str(len(val_filenames)))
```

Example image

```python
image = load_img("dataset/training/bulbasaur/00000001.png")
image = img_to_array(image)
print (image.shape)
array_to_img(image)
```

Initializing the image data generators for training and validation. 

```python
# Generate training tata from the images in a folder
train_batch_size = 64
train_datagen = ImageDataGenerator(rescale=1./255)
train_generator = train_datagen.flow_from_directory(
    'dataset/training/',
    target_size=(224, 224),
    batch_size=train_batch_size,
    class_mode='categorical'
    )


# Generate validation tata from the images in a folder
val_batch_size = 10
val_datagen = ImageDataGenerator(rescale=1./255)
val_generator = val_datagen.flow_from_directory(
    'dataset/validation/',
    target_size=(224, 224),
    batch_size=val_batch_size,
    class_mode='categorical'
    )
```

Initializing the model. I will be using 3 conv layers with batch normalizationa and max pooling layers followed by a fully connected layer and finally a soft max layer. 

```python
chanDim=-1 #Since its not channel_first
model = Sequential()
model.add(Conv2D(16, (3, 3), padding="same",input_shape=(224,224,3)))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))
model.add(MaxPooling2D(pool_size=(3, 3)))
model.add(Dropout(0.25))
model.add(Conv2D(32, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), padding="same"))
model.add(Activation("relu"))
model.add(BatchNormalization(axis=chanDim))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256))
model.add(Activation("relu"))
model.add(BatchNormalization())
model.add(Dropout(0.5))
# softmax classifier
model.add(Dense(4))
model.add(Activation("softmax"))

#compile the model using Adam optimizer.
model.compile(loss="categorical_crossentropy", optimizer="adam",metrics=["accuracy"])
```

```python
model.summary()
```

Running  the model for 100 epochs. 

```python
EPOCHS = 100
H = model.fit_generator(
    train_generator,
    steps_per_epoch=len(train_generator.filenames)//train_generator.batch_size,
    validation_data=val_generator,
    validation_steps=len(val_generator.filenames)//val_generator.batch_size,
    epochs=EPOCHS, verbose=1)
```

It seems to have done decently. Here is the plot of the losses from the training. 

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(H.history['loss']);
ax[0].set_title("Loss")
ax[1].plot(H.history['val_loss'])
ax[1].set_title("Validation Loss")
```

Let us test the model on some test images. 

```python
test_img = load_img("testing/pika_test1.jpg", target_size=(224, 224))
plt.imshow(test_img)
plt.axis("off")
test_img = img_to_array(test_img)
test_img = np.expand_dims(test_img, axis=0)
tmp = model.predict(test_img)
print ("This is : "+list(train_generator.class_indices.keys())[list(train_generator.class_indices.values()).index(np.argmax(tmp))])
```
